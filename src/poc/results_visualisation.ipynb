# Results Visualisation

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Rule Learning (Control Data)
To compare how our Doc2Vec model performs, we have chosen to use the performance of the previous approach as our baseline. This model has been evaluated using 10-fold cross validation.

```python
n_projects = 3

pitsA = (0.51, 0.80, 0.51)
pitsB = (0.62, 0.69, 0.64)
pitsC = (0.81, 0.78, 0.79)
pitsD = (0.46, 0.47, 0.46)
pitsE = (0.38, 0.60, 0.41)

fig, ax = plt.subplots(figsize=(8,6))

index = np.arange(n_projects)
bar_width = 0.15

pitsA = ax.bar(index, pitsA, bar_width, label='A')
pitsB = ax.bar(index + bar_width, pitsB, bar_width, label='B')
pitsC = ax.bar(index + 2*bar_width, pitsC, bar_width, label='C')
pitsD = ax.bar(index + 3*bar_width, pitsD, bar_width, label='D')
pitsE = ax.bar(index + 4*bar_width, pitsE, bar_width, label='E')


ax.set_xlabel('')
ax.set_ylabel('')
ax.set_title('10-Fold Cross Validation Rule Learning')
ax.set_xticks(index + 4*bar_width/2)
ax.set_xticklabels(('Recall', 'Precision', 'F-Measure'))
ax.legend()

fig.tight_layout()
plt.show()
fig.savefig('rl_evaluation_bar.png')
```

## Rule Learning vs. Doc2Vec
We will now compare how our model compares to rule learning when also tested with 10-fold cross-validation.

```python
n_projects = 5

rl_recall = (0.51, 0.62, 0.81, 0.46, 0.38)
d2v_recall = (0.53, 0.55, 0.84, 0.91, 0.57)

fig, ax = plt.subplots(figsize=(8,6))

index = np.arange(n_projects)
bar_width = 0.20

rects_rl = ax.bar(index, rl_recall, bar_width, label='rule learing')
rects_d2v = ax.bar(index + bar_width, d2v_recall, bar_width, label='doc2vec')


ax.set_xlabel('PITS Projects')
ax.set_ylabel('')
ax.set_title('Rule Learning vs. Doc2Vec Recall')
ax.set_xticks(index + bar_width/2)
ax.set_xticklabels(('A', 'B', 'C', 'D', 'E'))
ax.legend()

fig.tight_layout()
plt.show()
fig.savefig('rl_d2v_recall_bar.png')
```
