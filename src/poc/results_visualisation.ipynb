# Results Visualisation

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Rule Learning (Control Data)
To compare how our Doc2Vec model performs, we have chosen to use the performance of the previous approach as our baseline. This model has been evaluated using 10-fold cross validation.

```python
n_projects = 5

recall = (0.51, 0.62, 0.81, 0.46, 0.38)
precision = (0.80, 0.69, 0.78, 0.47, 0.60)
fmeasure = (0.51, 0.64, 0.79, 0.46, 0.41)

fig, ax = plt.subplots(figsize=(8,6))

index = np.arange(n_projects)
bar_width = 0.15

pitsA = ax.bar(index, recall, bar_width, label='recall')
pitsB = ax.bar(index + bar_width, precision, bar_width, label='precision')
pitsC = ax.bar(index + 2*bar_width, fmeasure, bar_width, label='f-measure')


ax.set_xlabel('PITS Projects')
ax.set_ylabel('')
ax.set_title('10-Fold Cross Validation Rule Learning')
ax.set_xticks(index + 2*bar_width/2)
ax.set_xticklabels(('A', 'B', 'C', 'D', 'E'))
ax.legend()

fig.tight_layout()
plt.show()
fig.savefig('rl_evaluation_bar.png')
```

## Doc2Vec Past Mixed with Percentage of Current
Starting with 0% of the current project used for training to increasing percentages of {20, 50, 80, 95}.

```python
percentages = (0.00, 0.20, 0.50, 0.80, 0.95)

pitsA = (0.33, 0.50, 0.50, 0.49, 0.41)
pitsB = (0.53, 0.58, 0.59, 0.63, 0.86)
pitsC = (0.84, 0.88, 0.90, 0.96, 1.00)
pitsD = (0.91, 0.98, 1.00, 1.00, 1.00)
pitsE = (0.56, 0.60, 0.55, 0.64, 0.65)
pitsF = (0.59, 0.50, 0.49, 0.54, 0.76)

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(percentages, pitsA, 'o-', label='A')
ax.plot(percentages, pitsB, 'o-', label='B')
ax.plot(percentages, pitsC, 'o-', label='C')
ax.plot(percentages, pitsD, 'o-', label='D')
ax.plot(percentages, pitsE, 'o-', label='E')
ax.plot(percentages, pitsF, 'o-', label='F')

ax.set_xlabel('% of current project used to train')
ax.set_ylabel('f-measure')
ax.set_title('D2V Past with Percentage of Current Project')
ax.legend(loc='lower center', ncol=6)

fig.tight_layout()
plt.show()
fig.savefig('d2v_percentage_f-measure_line.png')
```
